```python
import pandas as pd
import os

#Chemin vers le dossier contenant les fichiers
chemin_dossier = '/content'

#Liste pour stocker les DataFrames
dataframes = []
```

```python

#Parcourir tous les fichiers dans le dossier
for fichier in os.listdir(chemin_dossier):
    # Construire le chemin complet vers le fichier
    chemin_complet = os.path.join(chemin_dossier, fichier)

#Vérifier si le fichier est un CSV (vous pouvez ajuster cette condition selon vos besoins)
    if chemin_complet.endswith('.csv'):
        # Charger le fichier CSV avec séparation par tabulation
        df = pd.read_csv(chemin_complet, sep='\t')
```

```python
dataframes.append(df)

#Si vous souhaitez concaténer tous les DataFrames en un seul DataFrame
#Remarque : cela suppose que tous les fichiers ont la même structure de colonnes
donnees_concatenees = pd.concat(dataframes, ignore_index=True)

#Afficher les premières lignes du DataFrame concaténé pour vérification
donnees_concatenees.head()
```

```python
donnees_concatenees.to_csv('/content/Data.csv', index=False)
```
